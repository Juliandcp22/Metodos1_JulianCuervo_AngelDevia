```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python

x= np.linspace(-4,4,25)
y= x.copy()


h = 0.001
```

```python
def function(x,y):
  return (2*x*(1-(2**2/(x**2+y**2))))
```

```python
X,Y= np.meshgrid(x,y)
V = function(X,Y)
```

```python
#function(X,Y)
```

```python

def Dx(f,x,y,h):
    return (f(x+h,y)-f(x-h,y))/(2*h)
def Dy(f,x,y,h):
    return -1*(f(x,y+h)-f(x,y-h))/(2*h)
```

```python

N = 25
Ex = np.zeros((N,N))
Ey = np.zeros((N,N))

for i in range(N):
    for j in range(N):
        Ex[i,j] = Dx(function,x[i],y[j],h)
        Ey[i,j] = Dy(function,x[i],y[j],h)
```

```python
fig1 = plt.figure(figsize=(10,10))
ax = fig1.add_subplot(1,1,1)

tube= plt.Circle((0,0), 2, fill= False, color="r", linewidth=5)

for i in range(1,len(x)):
    for j in range(1,len(y)):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j],color='b')
ax.add_artist(tube)

```
