```python
def sucesion_de_fibonacci(n:int)->list:
   contador1=0
   contador2=1
   i=3
   suma=0
   sucesion=[0,1]
   if n==1:
       sucesion.remove(1)
   elif n==2:
       sucesion=[0,1]
   elif n>2:
       while i<=n:
       
           suma=contador1+contador2
           contador1=contador2
           contador2=suma
           sucesion.append(suma)
           i+=1
   
   return sucesion
```

```python
n=20
sucesion_de_fibonacci(n)
```

```python
def fibonacci(n:int)->int:  
  contador1=0
  contador2=1
  i=3
  suma=0
  sucesion=[1]
  if n==1:
       sucesion.remove(1)
  elif n==2:
       sucesion=[1]
  elif n>2:
       while i<=n:
       
           suma=contador1+contador2
           contador1=contador2
           contador2=suma
           sucesion.append(suma)
           i+=1
  return sucesion 
n=20
fibonacci(n)
```
