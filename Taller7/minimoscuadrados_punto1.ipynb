```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A= np.array([[2,-1],
            [1,2],
            [1,1]])
b= np.array([2,1,4])

AT = np.dot(A.T,A)

x1,y1= np.linalg.inv(AT)@(A.T@b)
```

```python
def recta1(x):
  return 2*x-2
```

```python
def recta2(x):
  return (-x+1)/2
```

```python
def recta3(x):
  return -x+4
```

```python
x=np.linspace(-5,5,333)
y=np.copy(x)
```

```python
r1=recta1(x)
plt.plot(x,r1)
r2=recta2(x)
plt.plot(x,r2)
r3=recta3(x)
plt.plot(x,r3)
plt.scatter(x1,y1,c="purple")
plt.xlim(left=-5,right=5)
plt.ylim(bottom=-5,top=5)
```

```python
def diferencia(x,y):
  return np.sqrt(6*x**2+2*x*y-18*x+6*y**2-8*y+21)
```

```python
z1=diferencia(x1,y1)

z1
```

```python
from matplotlib.axis import Axis
xs,ys = np.meshgrid(x,y, sparse=True)
zs = diferencia(xs,ys)
fig = plt.figure(figsize=(30,9))
ax = fig.add_subplot(1,2,1,projection='3d')
surf = ax.plot_surface(xs, ys, zs, cmap="coolwarm",
                       linewidth=2, antialiased=False)
z1=diferencia(x1,y1)
ax.set_zlabel('d(x,y)')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.plot(x,recta1(x))
ax.plot(x,recta2(x))
ax.plot(x,recta3(x))

ax.view_init(20, 40)
ax.scatter(x1,y1,z1,c="b",s=60)
```
